```python
import pandas as pd
import sys

model_preds = pd.read_csv("../csv/model_preds.csv")
gt = pd.read_csv("../csv/gt.csv")
```

```python
model_tops_list = model_preds['Achieved TOps'].tolist()
gt_tops_list = gt['tops'].tolist()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(len(model_tops_list))
width = 0.35

plt.bar(x - width/2, model_tops_list, width, label='Model Preds')
plt.bar(x + width/2, gt_tops_list, width, label='GT')
plt.title("TOPs")
plt.legend()
plt.show()
```

```python
model_oi_list = model_preds['Operational Intensity (DRAM)'].tolist()
gt_oi_list = gt['oi'].tolist()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(len(model_tops_list))
width = 0.35

plt.bar(x - width/2, model_oi_list, width, label='Model Preds')
plt.bar(x + width/2, gt_oi_list, width, label='GT')
plt.title("OI")
plt.legend()
plt.show()
```
